```python
import requests
import pandas as pd

url = "https://world.openfoodfacts.org/cgi/search.pl"
params = {
    "search_simple": 1,
    "action": "process",
    "json": 1,
    "page_size": 1000,
    "page": 1,
}

all_products = []

for page in range(1, 11):  # 1000 * 10 = 10,000 items
    print(f" Fetching page {page}")
    params["page"] = page
    response = requests.get(url, params=params)
    data = response.json()
    all_products.extend(data["products"])

df = pd.DataFrame(all_products)
df.to_csv("openfood_10k.csv", index=False)
print(" Saved 10,000 food items from OpenFoodFacts")

#scraping through the usage of api is preferred for large datasets.
```

```python
columns_to_remove = [
    'sources', 'sources_fields', 'taxonomies_enhancer_tags', 
    'weighters_tags', 'completed_t', 'grades', 'scores',
    'carbon_footprint_from_known_ingredients_debug',
    'serving_size_imported', 'allergens_imported'
]
df.drop(columns=columns_to_remove, inplace=True, errors='ignore')
#errors=ignore incase we have already removed it
```

```python
df['additives_n'] = df['additives_n'].fillna(0).astype(int)
```

```python
import ast

list_columns = ['additives_tags', 'additives_original_tags', '_keywords', 'added_countries_tags']

for col in list_columns:
    df[col] = df[col].apply(lambda x: x if isinstance(x, list) else [] if pd.isna(x) else ast.literal_eval(x) if isinstance(x, str) else [])
```

```python
for col in list_columns:
    print(col, df[col].apply(type).value_counts())
#sanity check
```

```python
constant_cols = []

for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, list)).any():
        continue  # skip columns that contain lists
    try:
        if df[col].nunique() <= 1:
            constant_cols.append(col)
    except TypeError:
        pass  # just in case something else is unhashable

df.drop(columns=constant_cols, inplace=True)
print(f"Dropped {len(constant_cols)} constant columns.")

```

```python
# Drop columns with more than 90% missing values
threshold = 0.9
df = df.loc[:, df.isnull().mean() < threshold]
print(f"Shape after dropping mostly-empty columns: {df.shape}")
```

```python
# Drop rows with more than 90% NaNs
df = df[df.isnull().mean(axis=1) < threshold]
print(f"Shape after dropping mostly-empty rows: {df.shape}")
```

```python
constant_cols = []

for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, list)).any():
        continue  # skip list-like columns
    try:
        if df[col].nunique(dropna=True) <= 1:
            constant_cols.append(col)
    except TypeError:
        pass

df.drop(columns=constant_cols, inplace=True)
print(f"Dropped {len(constant_cols)} constant columns.")
```

```python
list_cols = [col for col in df.columns if df[col].apply(lambda x: isinstance(x, list)).any()]

for col in list_cols:
    df[col] = df[col].apply(lambda x: ', '.join(
        [str(i) for i in x]  
    ) if isinstance(x, list) else x)

```

```python
print("🧹 Cleaned DataFrame shape:", df.shape)
print("Remaining NaNs per column:\n", df.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# Assuming `nutriments` column contains dicts
nutri_df = pd.json_normalize(df['nutriments'])
df_combined = pd.concat([df, nutri_df], axis=1)

# Map NutriScore A-E to numeric scale
nutriscore_map = {'a': 5, 'b': 4, 'c': 3, 'd': 2, 'e': 1}
df_combined['nutriscore_num'] = df_combined['nutriscore_grade'].map(nutriscore_map)

# Correlation matrix
corr_cols = ['nutriscore_num', 'nova_group', 'salt_100g', 'sugars_100g', 'fat_100g', 'proteins_100g', 'fiber_100g']
correlation = df_combined[corr_cols].corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix – Nutrition Metrics")
plt.show()
```

```python
sns.countplot(data=df_combined, x='nutriscore_grade', order=['a','b','c','d','e'], palette='Set2')
plt.title("NutriScore Distribution")
plt.show()
```
